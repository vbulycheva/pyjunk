```python
import pandas as pd
import numpy as np
coin_data = pd.read_csv('./data/victoria.csv')
copper_data = pd.read_csv('./data/bp8103_copper.csv', header = 20)

copper_data['Energy, keV'] = 20.0813 * copper_data['Channel#']/1000


import matplotlib.pyplot as plt

en = copper_data['Energy, keV']

intn = copper_data['Intensity']


plt.plot(en, intn,)
plt.xlabel("Energy, keV")
plt.ylabel('Counts')
plt.title('Copper, linear')
plt.grid (True)

plt.savefig('./out/cop_lin.png', dpi = 1000)

plt.semilogy(en, intn)
plt.title('Copper, log')
plt.xlabel('Energy, keV')
plt.ylabel('Counts')
plt.grid (True)
plt.savefig('./out/cop_log.png', dpi = 1000)
plt.show()
```
